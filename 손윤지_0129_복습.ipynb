#데이터 병합 방식

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                    'B': [4,5,6]})
df2 = pd.DataFrame({'A': [7,8,9],
                    'B': [10,11,12]})
```

```python
df1
```

#merge 함수수

left : 병합할 첫 번째 데이터 프레임, 왼쪽 df

right: 병합할 두 번쨰 데이터 프레임, 오른쪽 df

on : 조인 키로 사용할 컬럼 목록록, 여기에 입력된 컬럼은 꼭 두 데이터프레임에 모두 공통으로 포함 되어야 함!

left_on : 왼쪽 데이터프레임에서 조인 키로 사용할 컬럼 목록

right_on : 오른쪽 데이터프레임에서 조인 키로 사용할 컬럼 목록

left_index : 왼쪽 데이터프레임에서 조인 기준으로 인덱스를 사용할지

right_index: 오른쪽 데이터프레임에서 조인 기준으로 인덱스 사용할지

how: 조인 방법을 결정 ( 디폴트는 inner)


```python
df1 = pd.DataFrame({'A':[1,2,3,4],
                   'B':[5,6,7,8]})

df2 = pd.DataFrame({'A':[1,2,3,4],
                   'C':[10,11,12,13]})
```

```python
display(df1)
display(df2)
```

#on 인자 사용하기기

```python
pd.merge(df1, df2, on='A')
```

```python
pd.merge(df1, df2, on='B') #A처럼 중복이 아니라서서
```

#left_on, right_on

```python
df1 = pd.DataFrame({'A':[1,2,3,4],
                   'B':[5,6,7,8],
                   'D':[0,0,0,3]})

df2 = pd.DataFrame({'a':[1,2,3,4],
                   'B':[5,6,7,8],
                   'E':[1,1,1,0]})
```

```python
pd.merge(df1, df2, left_on=['A'],right_on=['a'])
```

```python
pd.merge(df1, df2, left_on=['A','B'],right_on=['a','B']) #B로 맞춰감감
```

#left_index, right_index

```python
df1 = pd.DataFrame({'A':[1,2,3,4],
                   'D':[0,0,0,3]})

df2 = pd.DataFrame({'A':[1,2,3,4],
                   'E':[1,1,1,0]})
```

```python
#인덱스 설정할 수 있음
#인덱스를 설정하는 방법은 컬럼.set_index('컬럼', inplace=True)
#인덱스를 리셋하는 방법은 컬럼.reset_index('컬럼', inplace=True)

df1.set_index('A',inplace=True) #A로 인덱스 설정정
```

```python
df1 
```

```python
pd.merge(df1, df2, left_index = True, right_on='A')
```

```python
pd.merge(df1, df2, left_index = True ,right_on='E')
```

#How

디폴트 inner : 교집합

right : 오른쪽 데이터프레임 조인 키 병합

left : 왼쪽 데이터프레임을 조인 키 병합

outer : 합집합


```python
df1 = pd.DataFrame({'A':[1,2,3,4],
                   'D':[0,0,0,3]})

df2 = pd.DataFrame({'A':[1,2,7,4],
                   'E':[1,1,1,0]})
```

```python
pd.merge(df1,df2, on='A', how='inner') #교집합 개념 그래서124
```

```python
pd.merge(df1,df2, on='A')
```

```python
pd.merge(df1,df2, on='A', how='left') #겹치는 것
```

```python
pd.merge(df1,df2, on='A', how='right')
```

```python
pd.merge(df1,df2, on='A', how='outer') #전체
```
