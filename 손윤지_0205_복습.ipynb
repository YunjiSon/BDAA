```python
pd.melt(pew, id_vars='religion') #지정한 열의 데이터를 모두 행으로 정리해 준다. 
#사용하는 특정한 경우가 있고, 그때 사용해서 데이터 전처리를 수월하게 할 수 있다
#id_vars: 위치를 그대로 유지하는 열의 이름
```
