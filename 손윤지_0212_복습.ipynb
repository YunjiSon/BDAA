##시계열 데이터
정상시계열, 비정상시계열

비정상시계열 주식가격, 기온, 매출, 주문량 다양한 데이터

비정상시계열 -> 계절성, 주기성

비정상시계열을 -> 정상시계열로 만들어서 학습하고 시계열 분석을 진행한다.



```python
import datetime
```

```python
date= datetime.date(year=2023, month=2, day=12)
time = datetime.time(hour=23, minute= 32, second= 5, microsecond=1234)
```

```python
time
```

```python
dt = datetime.datetime(year=2023, month=2, day=12,hour=23, minute= 32, second= 5, microsecond=1234)
```

```python
dt
```

```python
#timedelta

td = datetime.timedelta(days= 5, hours=5)
```

```python
td
```

```python
#timedelta를 가지고 계산할 수 있다. 시계열적계산

print(date+td)
print(dt+td)
```

##crime 데이터를 인덱스로 변경하자!
인덱스를 시계열 데이터로 만들기!

```python
crime_df = crime.set_index('REPORTED_DATE')
```

```python
crime_df
```

```python
#시계열 데이터를 인덱스로 변경 후에는, 슬라이싱, 원하는 요일, 연도 등 자유롭게 추출이 가능하다
crime_df.loc['2017']
```

```python
crime_df.loc['2017-09-12']
```

```python
crime_df.loc['Dec 2015']
```

```python
crime_df.loc['2017'].sort_index()
```

```python
crime_df.loc['2017-01-01':'2017-01-12'].sort_index()
```

##Between_time

```python
crime_df.between_time('10:00','15:00').sort_index()
```

```python
#원하는 시간만 추출
crime_df_sr= crime_df.at_time('17:12').sort_index()
```

```python
crime_df_sr
```

```python
crime_df_sr.first('5D')
```

##resample

```python
#월별, 일별, 주별 값을 계산

crime_df.resample('Y').sum()
```

```python
crime_df.resample('Y').max()
```
