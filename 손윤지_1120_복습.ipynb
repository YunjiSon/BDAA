집단별 요약 (groupby)

.groupby('컬럼') # 해당컬럼을 묶어서 보여준다

.agg(변수 = ('컬럼','원하는 통계치')

```python
grouped = mpg.groupby('manufacturer')
grouped
```

```python
#각종 계산이 용이
mpg.groupby('manufacturer').agg(sum_cty = ('cty','sum'))
```

```python
##여러 개의 변수 한꺼번에 볼 수 있음.

mpg.groupby('manufacturer').agg(mean_cty = ('cty','mean'),
                          cty_sum = ('cty','sum'),
                          cty_median = ('cty','median'),
                          n =('manufacturer','count'))
```

```python
mpg.groupby('manufacturer').sum()
```

데이터 병합하고 합치기

```python
a = pd.DataFrame({'id':[1,2,3,4,5],
                 'midterm':[50,60,70,80,90]})

b = pd.DataFrame({'id':[1,2,3,4,5],
                 'final':[40,60,80,30,80]})
```

```python
print(a)
display(a)
display(b)
```

```python
total = pd.merge(a, b, how='inner') #inner 는 교집합
total
```

```python
###concat : 데이터프레임을 말그대로 물리적으로 이어 붙여주는 함수
pd.concat([a,b],axis=0) #0은 행이고 1은 열이다
```
