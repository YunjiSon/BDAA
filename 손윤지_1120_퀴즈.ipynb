2. 현대, 토요타, 포드 중 고속도로 평균연비가 가장 좋은 회사의 차는?

```python
mpg_hwy = mpg.groupby('manufacturer').agg(mean_hwy = ('hwy','mean'))
```

```python
mpg_hwy3 = mpg_hwy.query('manufacturer=="hyundai"| manufacturer == "toyota" | manufacturer == "ford"')
mpg_hwy3
```

```python
mpg_hwy3.sort_values('mean_hwy', ascending=False)
```

3. 평균 제조년도가 오래된 순서로 자동차 모델 나열하기

```python
model_year = mpg.groupby('model').agg(year_mean = ('year','mean'))
model_year
```

```python
model_year.sort_values('year_mean')
```

4. 배기량이 2이상 2.5이하인 자동차와 5 이상인 자동차의 도시 연비 평균

```python
mpg.query('displ>=2 & displ<=2.5')
```

```python
mpg.query('displ>=2 & displ<=2.5')['cty'].mean()
```
